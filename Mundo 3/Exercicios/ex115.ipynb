### 115a: Vamos criar um menu em Python

```python
def lerInt(msg):
    while True:
        try:
            n = int(input(msg))
        except (TypeError, ValueError):
            print('\033[31mErro: Digite um valor inteiro válido!\033[m')
            continue
        except KeyboardInterrupt:
            print('\033[31mO usuário interrompeu o sistema.\033[m')
            return 3
        else:
            return n
```

```python
# Sem usar arquivos
def cadastrar(grupo):
    while True:
        try:
            nome = str(input('Digete o Nome: '))
            idade = int(input('Digite a Idade: '))
            pessoa = [nome, idade]
            grupo.append(pessoa)
        except (TypeError, ValueError):
            print('\033[31mErro: Digite um valor inteiro válido!\033[m')
            continue
        except KeyboardInterrupt:
            print('\033[31mO usuário interrompeu o sistema.\033[m')
            return 0
        else:
            print('Cadastro Realizado com sucesso!')
            return grupo[:]
    
def printGrupo(grupo):
    if len(grupo) == 0:
        print('\033[31mNenhuma pessoa cadastrada!\033[m')
    else:
        print(f'{"NOME":<15}{"IDADE":>6}')
        for pos in range(0, len(grupo)):
            print(f'{grupo[pos][0]:<15}{grupo[pos][1]:^6}')
```

```python
def menu(*msg):
    """
    -> Função para criar um escopo de menu
    param msg: lista de itens no menu, 0 primeiro deve ser o titúlo do menu
    return: A quantidade de itens do menu
    """
    tam = len(max(msg)) + 6
    print('\033[34m-'*(tam))
    print(f'{msg[0]}'.center(tam))
    print('-'*(tam),'\033[m')
    for i in range(1, len(msg)):
        print(f'\033[32m{i}\033[m - \033[34m{msg[i]}\033[m')
    if len(msg) > 1:
        print('\033[34m-\033[m'*(tam))
    return len(msg) -1
    
```

### 115b: Vamos ver como fazer acesso a arquivos usando o Python

```python
def arquivoExiste(nome):
    try:
        a = open(arq, 'rt')
        a.close()
    except FileNotFoundError:
        return False
    else:
        return True
        
        
def criarArquivo(nome):
    try:
        arquivo = open(nome, 'wt+')
        arquivo.close()
    except:
        print('Problema na criação do arquivo.')
    else:
        print(f'Arquivo {arquivo.name} criado com sucesso!')
        
    
def lerArquivo(nome):
    try:
        arquivo = open(nome, 'rt')
    except:
        print('Erro na leitura do arquivo!')
    else:
        menu('PESSOAS CADASTRADAS')        
        lista = arquivo.readlines()
        print(f'{"NOME":<15}{"IDADE":>6}')
        if not len(lista):
            print('\033[31mNenhuma pessoa cadastrada!\033[m')
        for item in lista:
            dado = item.split(';')
            print(f'{dado[0]:<15}{dado[1][:-1:]:>5} anos')
    finally:
        arquivo.close()

def escreverArquivo(arq, nome, idade):
    try:
        arquivo = open(arq, "a")
    except:
        print('Erro ao ler arquivo!')
    else:
        pessoa = f'{nome};{idade}\n'
        arquivo.writelines(pessoa)
        print('Cadastro Realizado com sucesso!')
    finally:
        arquivo.close()
```
